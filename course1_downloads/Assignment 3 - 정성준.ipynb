````python
def answer_one():
    import pandas as pd
    #
    def energy():
        import pandas as pd
        import numpy as np
        #readdata
        energy = pd.read_excel('Energy Indicators.xls')
        energy = energy.iloc[16:243,[1,3,4,5]]
        columnsnames = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
        for i in range(len(energy.columns)):
            energy = energy.rename(columns = {energy.columns[i]:columnsnames[i]})
        
        #remove any digits in the contry column
        chars_notinnums = []    
        for num in range(len(energy['Country'])):
            chars = list(energy['Country'].iloc[num])
            chars_notinnum = [char for char in chars if not char.isdigit()]
            chars_notinnums.append(''.join(chars_notinnum))
        energy['Country'] = chars_notinnums
        
        #replace all entries with '...' with np.NaN
        for i in range(len(energy.columns)):
            energy.iloc[:,i][energy.iloc[:,i] == '...'] = np.NaN
        #conver 'Energy Supply' values to giga_juols
        energy['Energy Supply'] *= 1000000
        #replace each country in dict's keys with the correspoinding dicts values
        dicts = {"Republic of Korea": "South Korea",
                "United States of America": "United States",
               "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
               "China, Hong Kong Special Administrative Region": "Hong Kong"}
        energy['Country'] = energy['Country'].replace(dicts)
        #remove the part after'(' form all countries that have parenthesis
        for i in range(len(energy)):
            energy.iloc[i,0]=energy.iloc[i,0].partition('(')[0]
            
        energy['Country'] = energy['Country'].replace({'Iran ' : 'Iran'})
        
        
        for i in range(1,len(energy.columns)):
            energy.iloc[:,i]=energy.iloc[:,i].astype('float64')
        energy = energy.reset_index()
        
        return energy
    
    def gdp():
        
        import pandas as pd
        #read data
        GDP = pd.read_csv('world_bank.csv',header = 4)
        #replace each country in dicts's keys with the corresponding dicts values
        dicts = {"Korea, Rep.": "South Korea",
                 "Iran, Islamic Rep.": "Iran",
                 "Hong Kong SAR, China": "Hong Kong"}
        GDP['Country Name'] = GDP['Country Name'].replace(dicts)
        #change column name 'Country Name' to 'Country'
        GDP = GDP.rename(columns = {'Country Name':'Country'})
        #keep only the column['Country', '2006', '2007', '2008','2009','2010', '2011','2012','2013','2014','2015']
        GDP = GDP[['Country', '2006', '2007', '2008','2009','2010', '2011','2012','2013','2014','2015']]
        return GDP
    def scimen():
        import pandas as pd
        ScimEn = pd.read_excel('scimagojr-3.xlsx')
        return ScimEn
    energy = energy()
    GDP = gdp()
    ScimEn = scimen()
    tdf = pd.merge(energy,ScimEn,how = 'right', left_on = 'Country', right_on='Country')
    tdf = pd.merge(GDP,tdf, how = 'right',left_on = 'Country', right_on='Country')
    res = tdf[tdf['Rank'] <= 15]
    res = res.set_index('Country')
    del res['index']
    return res
````

### Question 11 (6.6%)
Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia',

```

*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python
def answer_eleven():
    import pandas as pd
    Top15 = answer_one()
    country = Top15.index.tolist()
    Top15.reset_index(inplace=True)
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    Top15['Country'].replace(ContinentDict,inplace=True)
    Top15.rename(columns = {'Country':'Continent'},inplace=True)
    Top15['Country'] = country
    Top15_new = pd.DataFrame()
    Top15_new['size'] = Top15[['Continent','2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015']].set_index('Continent').groupby(level = 0).count().iloc[:,0]
    Top15_new['sum'] = Top15[['Continent','2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015']].set_index('Continent').groupby(level = 0).sum().iloc[:,0]
    Top15_new['mean'] = Top15[['Continent','2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015']].set_index('Continent').groupby(level = 0).mean().iloc[:,0]
    Top15_new['std'] = Top15[['Continent','2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015']].set_index('Continent').groupby(level = 0).std().iloc[:,0]
    Top15_new = Top15_new.astype('float64')
    answer = Top15_new
    return answer
````

### Question 12 (6.6%)
Cut % Renewable into 5 bins. Group Top15 by the Continent, as well as these new % Renewable bins. How many countries are in each of these groups?

*This function should return a __Series__ with a MultiIndex of `Continent`, then the bins for `% Renewable`. Do not include groups with no countries.*

````python
def answer_twelve():

    Top15 = answer_one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    import pandas as pd
    Top15['Continent'] = Top15.index
    Top15['Continent'] = Top15['Continent'].replace(ContinentDict)
    Top15['cut%'] = pd.cut(Top15['% Renewable'],5)
    a = Top15.groupby(by = ['Continent']).count().iloc[:,0]
    a=pd.DataFrame(a)
    answer = a.loc[a['2006'].isna() != True,'2006']
    
    return answer
````
